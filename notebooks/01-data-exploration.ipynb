```python
import pandas as pd
import numpy as np
from glob import glob
import cv2
import matplotlib.pyplot as plt
```

### Dataset

```python
train_images = glob("../../data/images/train/*.jpg")
valid_images = glob("../../data/images/valid/*.jpg")
train_labels = glob("../../data/labels/train/*.png")
valid_labels = glob("../../data/labels/valid/*.png")
```

```python
print(f'Training Dataset:   {len(train_images)} images and {len(train_labels)} labels')
print(f'Validation Dataset: {len(valid_images)} images and {len(valid_labels)} labels')
```

### Sample

```python
def plot(index):
    global train_images, train_labels
    name = train_images[index].split("/")[-1].split(".")[0]
    image = train_images[index]
    label = train_labels[index]
    image = cv2.imread(image)
    label = cv2.imread(label)
    plt.imshow(image[...,::-1])
    plt.imshow(label, alpha=0.55)
    plt.title(name)
    plt.show()
    return None
```

```python
plot(0)
```

```python
plot(1)
```

```python
plot(2)
```
